```python
import sqlite3
import pandas as pd
```

```python
# read db file
conn = sqlite3.connect('data/dallas-ois.sqlite')
cursor = conn.cursor()
```

```python
# get the table name
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall()
```

```python
# read the database as dataframe
incidents = pd.read_sql("SELECT * FROM incidents",conn)
officers = pd.read_sql("SELECT * FROM officers",conn)
subjects = pd.read_sql("SELECT * FROM subjects",conn)
officers.head()
```

```python
# join subject and incidents

subject_incidents = pd.read_sql('SELECT * FROM subjects JOIN incidents ON subjects.case_number=incidents.case_number',conn)
subject_incidents.head()
```

```python
# map the `subject_status`
mapping = {'Deceased':1000, 'Injured':500, 'Shoot and Miss': 250}
dataset['subject_statuses'] = dataset['subject_statuses'].apply(lambda x: mapping.get(x,100))
dataset.head()
```

**Visulize**

```python
# s = size of features, c = features color, alpha = transparency of the dots
dataset.plot.scatter(x='longitude', y='latitude', s='subject_statuses', c='race',alpha=0.25)
```
