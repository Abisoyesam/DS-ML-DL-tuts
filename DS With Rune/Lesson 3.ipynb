### Bar Chart

```python
# bar plot

data['NGA'].plot.bar(figsize=(20,6))
```

```python
data[['DNK','USA']].plot.bar(figsize=(20,6))
```

**Plot a range of data**

`.loc[from:to]` apply this on the DataFrame to get a range (both inclusive)

```python
# loc -> location

data[['DNK','USA']].loc[2000:].plot.bar(figsize=(20,6));
```

### Histogram

* `.plot.hist()` Create a histogram
* `bins=<number of bin>` Specify the number of bins in the histogram

```python
data['USA'].plot.hist(bins=15)
```

<hr>

### Pie Chart

* `.plot.pie()` Create a Pie Chart
* A few arguments are:
    * colors, labels, title, ylabel
    * autopct='%1.1f%%' sets percentage on chart

```python
df = pd.Series(data=[4, 8, 5], index=['Aspirant','Undergraduate','Post graduate'])
df.plot.pie(ylabel='')
```

```python
(data['USA'] < 20).value_counts().plot.pie(labels=['> 20', '< 20'],autopct='%1.1f%%')
```

## Data Presentation

Assuming we want to give a picture of how US CO2 per capita is compared to the rest of the world
* Let's take 2017 (as more recent data is incomplete)
* What is the mean, max, and min CO2 per capital in the world.

```python
data = pd.read_csv('./data/WorldCO2.csv', index_col=0)
data.head()
```

```python
data['USA'].describe()
```

```python
data.loc[2017].describe()
```

```python
data.loc[2017]['USA']
```

**How can we tell a story**
* US is above the mean
* US is not the max
* It is above 75%

```python
# some more advanced ploting
year = 2017

ax = data.loc[year].plot.hist(bins=15,facecolor='green')
ax.set_xlabel('CO2 per capita')
ax.set_ylabel('Number of countries')
ax.annotate("USA", xy=(15,5), xytext=(15,30), arrowprops=dict(arrowstyle="->",connectionstyle='arc3'));
```
