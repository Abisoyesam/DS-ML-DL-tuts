# Matplotlib

It is python library used for data visualization.

### Why Matplotlib?

* built on Numpy arrays and python
* integrate directly with pandas
* can create basic and advanced plots
* simple to use interface


```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB.

```python
x = [1,2,3,4]
y = [11,22,33,44]

plt.plot(x,y);
```

```python
# MATPLOTLIB 1st Object Oriented Method

fig = plt.figure() # to create a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd Method

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method (recommended)

fig, ax = plt.subplots()
ax.plot(x,y); # add data
```

```python
fig, ax = plt.subplots()
ax.plot(x,[50,100,200,250]); # add data
```

### Matplotlib example Workflow

```python
#0. Import matplotlib
%matplotlib inline

#1. Prepare data
x = [1,2,3,4,5]
y = [5,10,15,20,25]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width, height

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save the figure
fig.savefig("sampleplot.png")
```

### Bar Chart

```python
# make a plot from dictionary

no_of_txtbooks = {
    "Thermodynamic ": 8,
    "Fluid Mechanics ": 10,
    "Engineering Math ": 2,
    "Vibration ": 4
}

fig, ax = plt.subplots()
ax.bar(no_of_txtbooks.keys(), no_of_txtbooks.values())
ax.set(title="Engineering Library",
       xlabel="Textbook",
       ylabel="No. of textbook");
```

```python
# horizontal bar chart

fig, ax = plt.subplots()
ax.barh(list(no_of_txtbooks.keys()), list(no_of_txtbooks.values()));
```

### Histogram

```python
# Histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplot

```python
# Method one

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5));
```

```python
# plot to different axis

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5));
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(no_of_txtbooks.keys(), no_of_txtbooks.values());
ax4.hist(x);
```

```python
# Method two

fig, ax = plt.subplots(nrows=2, 
                       ncols=2,
                       figsize=(10,5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(no_of_txtbooks.keys(), no_of_txtbooks.values())
ax[1, 1].hist(x);
```

### Ploting from Pandas DataFrames

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
# remove last two zeros

car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
# add sale date column >> pd.date_range

car_sales["Sale Date"] = pd.date_range("10/11/2021", periods=len(car_sales))
car_sales
```

```python
# Cummulative sum >> astype(int) >> cumsum()

car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plot total sales

car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
# 'kind' parameter to specify scatter plot

car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
heart_diseases = pd.read_csv("heart-disease.csv")
heart_diseases
```

```python
heart_diseases.head(10)
```

```python
# create a histogram to show the distribution of the age

heart_diseases["age"].plot.hist();
```

```python
heart_diseases.plot.hist(figsize=(10,30), subplots=True);
```

```python
heart_diseases.head()
```

```python
over_50 = heart_diseases[heart_diseases["age"] > 50]

len(over_50)
```

```python
# pyplot method

over_50.plot(kind="scatter", x = "age", y = "chol", c="target");
```

```python
# OOP method mixed with pyplot

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax); 

# ax parameter helps with the labels
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax); 
ax.set_xlim([45,100])
```

```python
# OOP method from scratch

fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"], y= over_50["chol"], c=over_50["target"]);

#customise the plot 
ax.set(title="Heart diseases and cholestrol levels", xlabel="Age", ylabel="Cholestrol");

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
heart_diseases.head()
```

```python
# subplot of age, thalach, chol

fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True) # sharex means the same xlabel for subplots

# add data to axis
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target);

# customize ax0
ax0.set(title="Heart diseases and cholesterol levels",xlabel="Age", ylabel="Cholestrol");

# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50.chol.mean(), linestyle='--');


# plot axis1
scatter1 = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target);

# customize the plot
ax1.set(title="Heart diseases and Thalach level", ylabel="Max Heart Rate")
ax1.legend(*scatter1.legend_elements(), title="Target")

# add horizontal mean line
ax1.axhline(y=over_50.thalach.mean(), linestyle='--');

# add title to the figure
fig.suptitle("Heart Diseases Analysis", fontsize=16, fontweight="bold")
```

```python
# OOP method from scratch

fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"], y= over_50["chol"], c=over_50["target"], cmap="winter");

#customise the plot 
ax.set(title="Heart diseases and cholestrol levels", xlabel="Age", ylabel="Cholestrol");

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# subplot of age, thalach, chol

fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True) # sharex means the same xlabel for subplots

# add data to axis
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap="winter");

# customize ax0
ax0.set(title="Heart diseases and cholesterol levels",xlabel="Age", ylabel="Cholestrol");

# set limit
ax0.set_xlim([50,80])

# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50.chol.mean(), linestyle='--');


# plot axis1
scatter1 = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target, cmap="winter");

# customize the plot
ax1.set(title="Heart diseases and Thalach level", ylabel="Max Heart Rate")
ax1.legend(*scatter1.legend_elements(), title="Target")

# set limit
# ax1.set_xlim([50,80]) # N.B: not necessary if the subplot share same x-axis
ax1.set_ylim([60,200])

# add horizontal mean line
ax1.axhline(y=over_50.thalach.mean(), linestyle='--');

# add title to the figure
fig.suptitle("Heart Diseases Analysis", fontsize=16, fontweight="bold");
```
